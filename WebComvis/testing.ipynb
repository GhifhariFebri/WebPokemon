```python
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os

model = tf.compat.v1.keras.models.load_model('pokemon_classifier.h5')
image_base_dir = r"C:\Users\AGIF\Downloads\archive(1)\archive\Pokemon Dataset2\Pokemon Dataset1"

# List available directories in the base path
available_directories = os.listdir(image_base_dir)
class_names = available_directories

print(class_names)

# Path to the image file you want to make a prediction on
image_path = r"C:\Users\AGIF\Downloads\1200px-0727Incineroar.png" # Ganti dengan path gambar yang ingin Anda prediksi

# Load and preprocess the image
img = image.load_img(image_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0  # Normalisasi gambar

# Make predictions
predictions = model.predict(img_array)
predicted_class_index = np.argmax(predictions, axis=1)[0]
predicted_class_prob = np.max(predictions, axis=1)[0]
predicted_class_name = class_names[predicted_class_index]  # Pastikan class_names sesuai dengan model Anda

# Print the prediction
print(f'Predicted class: {predicted_class_name}')
print(f'Probability: {predicted_class_prob:.2f}')

# Visualize the image
plt.imshow(img)
plt.axis('off')
plt.title(f'Predicted class: {predicted_class_name}')
plt.show()
```

```python
import tensorflow as tf

# Load the model
model = tf.compat.v1.keras.models.load_model('pokemon_classifier.h5')
# Cari layer output terakhir
output_layer = model.layers[-1]

# Pastikan layer output adalah Dense dan menggunakan softmax activation
assert isinstance(output_layer, tf.keras.layers.Dense), "Last layer should be Dense"
assert output_layer.activation == tf.keras.activations.softmax, "Last layer should use softmax activation"
# Dapatkan jumlah kelas dari dimensi output layer
num_classes = output_layer.units
print(f"Number of classes in the model: {num_classes}")
```
